```python
import pandas as pd
data = pd.read_csv('web_ML_data_cleaned.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
import nltk
from nltk.tokenize import word_tokenize

string_columns = ['Title', 'Description', 'URL', 'keyword_final']

for col in string_columns:
    data[col] = data[col].apply(lambda x: word_tokenize(str(x)))

# Print the tokenized data
print(data[string_columns])

# Store the tokenized data back into the original data
data[string_columns] = data[string_columns]

# Print the updated DataFrame with tokenized data
data.head()
```

```python
#saving tokenized data
data.to_csv('web_ML_data_tokenized.csv', index=False)
```
